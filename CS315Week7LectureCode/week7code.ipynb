```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```python
# Load the Universal Sentence Encoder's TF Hub module
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

`embed` is: an instance of a class that is loading a saved model for the user.

`embed` expects a list of words, thus, even if we want the embedding of one single word, we put it in a list:

```python
# get the word embedding of a single word
embed(["apple"])
```

```python
words = ['king', 'queen', 'radio', 'TV', 'bike', 'car', 'Boston', 'London', 'lake', 'river']

embeddings = embed(words)
embeddings.shape
```
